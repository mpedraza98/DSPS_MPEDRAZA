What you know: The ratio of the area of the circle to the area of the square is Ï€ / 4.

```
def square(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
  
```

```
def circle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
  
```

```
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
```

```
def isin(x, y, l):
  d = np.sqrt(x**2+y**2)
  return d < l
```

```
np.random.seed(123)
```

```
# Lets generate a set of points using an uniform disbtribution for both x and y

x = np.random.uniform(-1,1,10000)
y = np.random.uniform(-1,1,10000)

pi = isin(x,y,1).sum()/10000.
```
